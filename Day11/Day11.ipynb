```python
#Part 1

import math
lines = open('11_big.txt', 'r').readlines()

class Monkey:
    def __init__(self, name, items, operator, number, divisor, first, second):
        self.name = name
        self.items = items
        self.operator = operator
        self.divisor = divisor
        self.first = first
        self.second = second
        self.number = number
        self.inspect = 0
        if self.number == 'old':
            self.operator = 'square'
    
    def operation(self):
        give = []
        for i in self.items:
            if self.operator == '+':
                new = i + int(self.number)
            if self.operator == '*':
                new = i * int(self.number)
            if self.operator == 'square':
                new = i * i
            new = math.floor(new/3)
            if new % self.divisor == 0:
                give.append((self.first, new))
            else:
                give.append((self.second, new))
            self.inspect += 1
        self.items = []
        return give

name = ''
items = []
operator = first = second = ''
number = divisor = 0
monkeys = []

for line in lines:
    if 'Monkey' in line:
        items = []
        name = line.strip(':\n')
    if 'Starting' in line:
        line = line.strip('\tStarting items: ').strip('\n')
        numbers = line.split(', ')
        for i in numbers:
            items.append(int(i))
    if 'Operation' in line:
        words = line.strip('\n').split(' ')
        operator = words[6]
        number = words[7]
    if 'Test' in line:
        divisor = int(line.strip('\tTest: divisible by ').strip('\n'))
    if 'true' in line:
        words = line.strip('\n').split(' ')
        first = words[9]
    if 'false' in line:
        words = line.strip('\n').split(' ')
        second = words[9]
        monkey = Monkey(name, items, operator, number, divisor, first, second)
        monkeys.append(monkey)

for n in range(20):
    for m in monkeys:
        for i in m.operation():
            who, what = i
            who = int(who)
            monkeys[who].items.append(what)

all = []            
for m in monkeys:
    all.append(m.inspect)

all.sort(reverse=True)
monkey_business = all[0] * all[1]

print(monkey_business)
```
